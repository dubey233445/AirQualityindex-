```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("datasets\clean_data.csv")
```

```python
df.head()
```

```python
# Function to categorize PM2.5 values
def categorize_pm(pm_value):
    if pm_value <= 30:
        return 1
    elif pm_value <= 60:
        return 2
    elif pm_value <= 90:
        return 3
    elif pm_value <= 120:
        return 4
    elif pm_value <= 250:
        return 5
    else:
        return 6

# Apply the function to create the 'category' column
df['PM2.5_category'] = df['PM2.5'].apply(categorize_pm)
```

```python
# Function to categorize PM10 values
def categorize_pm10(pm_value):
    if pm_value <= 50:
        return 1
    elif pm_value <= 100:
        return 2
    elif pm_value <= 250:
        return 3
    elif pm_value <= 350:
        return 4
    elif pm_value <= 430:
        return 5
    else:
        return 6

# Apply the function to create the 'category' column
df['PM10_category'] = df['PM10'].apply(categorize_pm10)
```

```python
# Function to categorize NO2 values
def categorize_NO2(no2_value):
    if no2_value <= 40:
        return 1
    elif no2_value <= 80:
        return 2
    elif no2_value <= 180:
        return 3
    elif no2_value <= 280:
        return 4
    elif no2_value <= 400:
        return 5
    else:
        return 6

# Apply the function to create the 'category' column
df['NO2_category'] = df['NO2'].apply(categorize_NO2)
```

```python
# Function to categorize CO values
def categorize_CO(co_value):
    if co_value <= 1:
        return 1
    elif co_value <= 2:
        return 2
    elif co_value <= 10:
        return 3
    elif co_value <= 17:
        return 4
    elif co_value <= 34:
        return 5
    else:
        return 6

# Apply the function to create the 'category' column
df['CO_category'] = df['CO'].apply(categorize_CO)
```

```python
# Function to categorize SO2 values
def categorize_SO2(so2_value):
    if so2_value <= 40:
        return 1
    elif so2_value <= 80:
        return 2
    elif so2_value <= 380:
        return 3
    elif so2_value <= 800:
        return 4
    elif so2_value <= 1600:
        return 5
    else:
        return 6

# Apply the function to create the 'category' column
df['SO2_category'] = df['SO2'].apply(categorize_SO2)
```
