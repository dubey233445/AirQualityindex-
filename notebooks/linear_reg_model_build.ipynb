```python
import pandas as pd
import os 
```

```python
file_path = os.path.abspath('datasets/city_day.csv')
df = pd.read_csv(file_path)
```

```python
df.info()
```

```python
null_values = df.isnull()

# If you want to count null values in each column, you can use sum()
null_counts = df.isnull().sum()

# If you want to see the total count of null values in the DataFrame, you can use sum() again
total_null_count = df.isnull().sum().sum()

# If you want to print out the null counts for each column
print("Null counts for each column:")
print(null_counts)

# If you want to print out the total count of null values
print("Total count of null values:", total_null_count)
```

```python
# Calculate the total number of entries in each column
total_entries_per_column = df.shape[0]

# Calculate the total number of missing values in each column
missing_values_per_column = df.isnull().sum()

# Calculate the percentage of missing values in each column
missing_percentage_per_column = (missing_values_per_column / total_entries_per_column) * 100

print("Percentage of missing values per column:")
print(missing_percentage_per_column)
```

```python
# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year component from the datetime objects
df['year'] = df['Date'].dt.year

# Get unique years
unique_years = df['year'].unique()

# Print the unique years
print("Unique years in the dataset:")
for year in unique_years:
    print(year)
```

```python
# Extract the unique cities from the DataFrame
unique_cities = df['City'].unique()

# Count the number of unique cities
num_unique_cities = len(unique_cities)

# Print the unique cities
print("Unique cities in the dataset:")
for city in unique_cities:
    print(city)

# Print the count of unique cities
print("Total unique cities:", num_unique_cities)
```

```python
# Convert the 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Define a custom function to fill missing values with the mean for numeric columns
def fill_numeric_mean(column):
    if pd.api.types.is_numeric_dtype(column.dtype):
        return column.fillna(column.mean())
    else:
        return column

# Apply the custom function to each column
df_filled = df.apply(fill_numeric_mean)

# Print the filled DataFrame
print(df_filled)
```

```python
df_filled
```

```python
df_filled.drop(columns=['AQI_Bucket','year'], inplace=True)
```

```python
#df_filled.to_csv('clean_data.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import joblib
```

```python
X = df_filled[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3']]
y = df_filled['AQI']
```

```python
from sklearn.preprocessing import StandardScaler

# Assuming 'X' is your feature matrix
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Initialize the linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
# Print coefficients
print("Coefficients:", model.coef_)
```

```python
# Calculate the R-squared value
r2 = r2_score(y_test, y_pred)
print("R-squared value:", r2)
```

```python
joblib.dump(model, 'models/linear_reg_model.pkl')
```
